```
def lista_ordenada(segundo_fator):
    lista_ordenada = dados[segundo_fator].unique()
    lista_ordenada.sort()
    return lista_ordenada
```

```
def gerar_boxplot(primeiro_fator, segundo_fator, tom=None, dados_analisar=dados, titulo_da_vertical="Y", titulo_da_horizontal="X"):
    plt.figure(figsize=(16,8))
    sns.boxplot(x=segundo_fator, y = primeiro_fator, hue=tom, data= dados_analisar, order=lista_ordenada(segundo_fator))
    plt.title("Boxplot de {} por {}".format(titulo_da_vertical, titulo_da_horizontal))
```

```
gerar_boxplot('NU_NOTA_TOTAL','Q025', tom='IN_TREINEIRO', dados_analisar=provas_sem_notas_zero, titulo_da_vertical="NOTA TOTAL", titulo_da_horizontal="quantidade de carros")
```

```
provas_entrada = ["NU_NOTA_REDACAO", "NU_NOTA_CH", "NU_NOTA_CN", "NU_NOTA_LC"]
prova_saida = "NU_NOTA_MT"
provas_sem_notas_zero = provas_sem_notas_zero[provas].dropna()

notas_entrada = provas_sem_notas_zero[provas_entrada]
notas_saida = provas_sem_notas_zero[prova_saida]
```

```
notas_entrada
```

```
x = notas_entrada
y = notas_saida
```

```
from sklearn.model_selection import train_test_split

SEED = 666

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25, random_state=SEED)
```

```
y_treino.shape
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
```

```
predicoes_matematica = modelo.predict(x_teste)
```

```
y_teste[:5]
```

```
plt.figure(figsize=(8,8))
sns.scatterplot(x=predicoes_matematica, y=y_teste)
plt.xlim(-50,1050)
plt.ylim(-50,1050)
```

```
plt.figure(figsize=(8,8))
sns.scatterplot(x=y_teste, y=y_teste - predicoes_matematica)
plt.xlim(-500,1050)
plt.ylim(-500,1050)
```
